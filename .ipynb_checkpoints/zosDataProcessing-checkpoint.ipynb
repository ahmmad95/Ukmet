```python
import pandas as pd
import numpy as np
import os
from datetime import datetime
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
%matplotlib inline
plt.rcParams["figure.figsize"] = (8,5)
import warnings
warnings.filterwarnings('ignore')
```

```python
Data=pd.read_excel(r'E:\MYS\Extracted Data\zos\Location5.xlsx')
Data=Data[Data['Sea Surface Height Above Geoid (m)']!=1.000000020040877e+20].reset_index(drop=True)
Data['date']=Data['date'].apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d %H:%M:%S'))
Data=Data[Data['date']<=datetime(2100, 12, 31)].reset_index(drop=True)
```

```python
df=Data[Data['projection']=='rcp45']
df.reset_index(drop=True,inplace=True)
#df['dff']=df.groupby(['model','projection','ensemble'])['Sea Surface Height Above Geoid (m)'].diff()
#df.dropna(inplace=True)
#df.reset_index(drop=True,inplace=True)
#df.set_index('date',inplace=True)
```

```python
for df1, df_g in df.groupby(['model','projection','ensemble']):
    index=df_g.index
    idxmin=min(index)
    df.loc[index,'change']=df.loc[index,'Sea Surface Height Above Geoid (m)']-df.loc[idxmin,'Sea Surface Height Above Geoid (m)']
    #print(min(index))
```
